### **Outliers Values**

Pada tahap ini, kita akan melihat nilai ekstrem (outliers) pada kolom yang bertipe data numerik. Hal ini dilakukan agar mendapatkan insight yang akurat sehingga dapat memperkuat analisis dan model yang akan kita bangun nantinya.

Akan tetapi memang tidak semua kolom perlu kita tangani, kita harus tetap melihat dan memahami tujuan dari setiap kolom yang ada di dataset.

Dan terlihat kolom

* total_electricity_share

Dengan persentase 25%

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

## **Relationship/Hubungan**

Aktivitas:
Menganalisis hubungan antara energi terbarukan dan energi fosil serta hubungannya dengan waktu (tahun).

Tujuan:
Mengetahui pola hubungan antar variabel dan melihat tren perubahan energi.

Visualisasi:
Heatmap

```python
corr_data = df[[
    'year',
    'renewable_energy_percent',
    'fossil_fuel_energy_percent',
    'electricity_from_renewables_percent',
    'electricity_from_fossil_percent'
]].dropna()

corr_matrix = corr_data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap Korelasi Energi Terbarukan, Energi Fosil, dan Tahun')
plt.show()
```

Heatmap menunjukkan bahwa energi terbarukan dan energi fosil memiliki hubungan saling menggantikan. Peningkatan energi terbarukan cenderung diikuti penurunan energi fosil (korelasi negatif). Listrik dari energi terbarukan berkorelasi positif dengan total energi terbarukan dan sangat negatif dengan listrik dari energi fosil, menandakan pergeseran sumber listrik. Sementara itu, hubungan variabel energi dengan tahun relatif lemah, sehingga tren perubahan tidak selalu konsisten setiap tahun.

## **Outliers**

```python
columns_to_impute = ["total_electricity_share", "total_electricity_share"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

## **Construct Data**

```python
df = df.sort_values(["country", "year"])
df["renewable_growth"] = df.groupby("country")["renewable_energy_percent"].diff()
df["renewable_growth"] = df["renewable_growth"].fillna(0)
```

```python
def tren_perkembangan_renewable_energy(growth):
    if growth > 0:
        return 'Meningkat'
    elif growth < 0:
        return 'Menurun'
    else:
        return 'Stabil'

df['tren_perkembangan_renewable_energy'] = df['renewable_growth'].apply(tren_perkembangan_renewable_energy)
```

Disini Saya menambahkan 2 kolom baru yaitu renewable_growth	dan tren_perkembangan_renewable_energy untuk dapat mrlihat tren arah perkembangan energi terbarukan ini apakah meningkat, stabil atau menurun.
